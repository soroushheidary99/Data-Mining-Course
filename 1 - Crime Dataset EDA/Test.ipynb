Q3 - may be spliting the data set based on year ... for more reliable answers by a more gradual and consecutive factor like month of year or even dayz ....


for this purpose we need to create a feature who will destinctify the year, more precisely the dates
we will also use a portion of the dataframe and not all, as we only need the dates for now
later for some additoinal Qs we will replace the Dates in crime_df by what we will work on now 

```python
date_df = crime_df[['DATE_OCCURED', 'DATE_REPORTED']]
date_df.head()
```

```python
print('-'*10, 'null amounts on each column', '-'*10)
print(date_df.isnull().sum())
print('-'*10, 'type amounts on date occure', '-'*10)
print(date_df.DATE_OCCURED.apply(lambda x : type(x)).value_counts())
print('-'*10, 'type amounts on date report', '-'*10)
print(date_df.DATE_REPORTED.apply(lambda x : type(x)).value_counts())
```

```python
date_df.DATE_OCCURED.fillna(value=date_df.DATE_REPORTED, inplace=True)
print('-'*10, 'null amounts on each column', '-'*10)
print(date_df.isnull().sum())
```

```python
date_df['stamp_len1'] = date_df.DATE_OCCURED.apply(lambda x : len(x))
date_df['stamp_len2'] = date_df.DATE_REPORTED.apply(lambda x : len(x))

print('-'*100)
print('before fixing dates : \n', date_df.groupby(by='stamp_len2').DATE_REPORTED.first())

def DateFixer(x) : 
    if '-' in x : 
        y = x[:4]
        m = x[5:7]
        d = x[8:10]
        h = x[11:13]
        x = str(int(m)) + '/' + str(int(d)) + '/' + str(int(y)) + ' ' + str(int(h))
    else : 
        x = x[:-3]
    return x


date_df.DATE_REPORTED = date_df.DATE_REPORTED.apply(lambda x : DateFixer(x))
date_df.DATE_OCCURED = date_df.DATE_OCCURED.apply(lambda x : DateFixer(x))
date_df['stamp_len1'] = date_df.DATE_OCCURED.apply(lambda x : len(x))
date_df['stamp_len2'] = date_df.DATE_REPORTED.apply(lambda x : len(x))

print('-'*100)
print('after fixing dates : \n', date_df.groupby(by='stamp_len2').DATE_REPORTED.first())
```

```python
date_df['HOUR_OCCURED'] = date_df.DATE_OCCURED.apply(lambda x : x[x.index(' ') + 1:])
date_df['HOUR_REPORTED'] = date_df.DATE_REPORTED.apply(lambda x : x[x.index(' ') + 1:])

date_df['DATE_OCCURED'] = date_df.DATE_OCCURED.apply(lambda x : x[:x.index(' ')])
date_df['DATE_REPORTED'] = date_df.DATE_REPORTED.apply(lambda x : x[:x.index(' ')])

date_df.drop(['stamp_len1', 'stamp_len2'], axis=1, inplace=True)
date_df.head()
```

```python
date_df['DATE_OCCURED'] = pd.to_datetime(date_df['DATE_OCCURED'], infer_datetime_format=True)
date_df['DATE_REPORTED'] = pd.to_datetime(date_df['DATE_REPORTED'], infer_datetime_format=True)
date_df.head()
```

```python
date_df['DAYZ_TILL_REPORT'] = date_df.DATE_REPORTED - date_df.DATE_OCCURED
date_df['DAYZ_TILL_REPORT'] = date_df['DAYZ_TILL_REPORT'].apply(lambda x : int(str(x).split(' ')[0]))

date_df.HOUR_OCCURED = date_df.HOUR_OCCURED.astype(int)
date_df.HOUR_REPORTED = date_df.HOUR_REPORTED.astype(int)
date_df['HOURS_TILL_REPORT'] = date_df['HOUR_REPORTED'] - date_df['HOUR_OCCURED']
date_df['HOURS_TILL_REPORT'] = date_df.HOURS_TILL_REPORT.apply(lambda x : x if x >= 0 else 24 + x)

date_df['REPORT_DELAY_HOUR'] = date_df.DAYZ_TILL_REPORT * 24 + date_df.HOURS_TILL_REPORT

date_df = date_df[['DATE_OCCURED', 'DATE_REPORTED', 'REPORT_DELAY_HOUR', 'HOUR_OCCURED']]
date_df.head()
```

```python
date_df['MONTH_OCCURED'] = date_df['DATE_OCCURED'].apply(lambda x : str(x)[:7])
date_df['CRIME_TYPE'] = crime_df.CRIME_TYPE
date_df.head()
```

```python
def Q3_f1(x) : 
    choice = None 
    x = x.tolist()
    max_num = 0 
    for i in x : 
        if x.count(i) > max_num : 
            max_num = x.count(i)
            choice = i
    return choice


crime_trends_whole = date_df.groupby(by='MONTH_OCCURED').CRIME_TYPE.agg(lambda x : Q3_f1(x))
crime_trends = crime_trends_whole[391:]
```

```python
fig = plt.figure(figsize=(24, 2))

plot = sns.scatterplot(x=crime_trends.index ,y=crime_trends.tolist())
plot.set_xticks(range(len(crime_trends.index)))


ticks = []
for i in range(len(crime_trends.index)) : 
    if i%5 != 0 :
        ticks.append(None)
    else : 
        ticks.append(crime_trends.index[i])

plot.set_xticklabels(ticks)
```

```python
crimes = date_df.CRIME_TYPE.unique().tolist()

def Q3_f2(x) : 
    dic = {}
    for i in crimes : 
        dic[i] = 0
    x = x.tolist()
    for i in x : 
        dic[i] += 1
    return dic

crime_numbers_whole = date_df.groupby(by='MONTH_OCCURED').CRIME_TYPE.agg(lambda x : Q3_f2(x))
crime_numbers = crime_numbers_whole[391:]
```

```python
plt.figure(figsize=(14, 8), dpi=360)

for name in crimes : 
    plt.plot([crime_numbers[i][name] for i in crime_numbers.index.tolist()], label=name)

plt.xticks(range(len(crime_trends.index)), ticks)
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
```
